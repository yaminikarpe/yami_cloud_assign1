```python
yami_df4 = pd.read_sql('''SELECT OWNERUSERID, OWNERDISPLAYNAME,TITLE,BODY FROM yami_table_assign1 WHERE OWNERUSERID IN (87234,4883,9951,6068,89904,51816,49153,179736,95592,63051)''', conn)
display(yami_df4)
```

```python
yami_df4['yami_title_body'] = yami_df4[['title', 'body']].agg(' '.join, axis=1)
```

```python
vectorizer = TfidfVectorizer(stop_words='english', lowercase=True)
yami_list = list(yami_df4["owneruserid"].unique())
for entry in yami_list:
    yami_entry_posts_df=yami_df4[yami_df4["owneruserid"] == entry]
    blue = vectorizer.fit_transform(yami_entry_posts_df["yami_title_body"])
    yami_tfidf_df= pd.DataFrame(blue.toarray(),columns=vectorizer.get_feature_names())
    
    yami_sum = yami_tfidf_df.sum(axis=0)
    top_ten = yami_sum.nlargest(10)
    yami_top_ten_wordslist = list(top_ten.index)
    yami_tfidf_top_ten=yami_tfidf_df[yami_top_ten_wordslist]
    
    display("TF-IDF of the top 10 terms for the user : "+entry)
    display(yami_tfidf_top_ten)
```
